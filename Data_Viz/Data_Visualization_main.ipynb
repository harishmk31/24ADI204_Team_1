```python
df.columns = [
    "List_Sr_No",
    "Set_No",
    "Set_Code",
    "First_Name",
    "Surname",
    "Country",
    "State_Association",
    "Date_of_Birth",
    "Age",
    "Role",
    "Batting_Style",
    "Bowling_Style",
    "Test_Caps",
    "ODI_Caps",
    "T20_Caps",
    "IPL_Matches",
    "Previous_IPL_Teams",
    "Current_Team",
    "No_of_Matches",
    "Cap_Status",
    "Reserve_Price"
]

df.head()
```

```python
df = df.drop(index=[0, 1]).reset_index(drop=True)

print("Shape After Removing First Two Rows:", df.shape)
df.head()
```

```python
df["Cap_Status"] = df["Cap_Status"].astype(str)

# If Cap_Status contains numbers, shift them to No_of_Matches
mask = df["Cap_Status"].str.isnumeric()

df.loc[mask, "No_of_Matches"] = df.loc[mask, "Cap_Status"]
df.loc[mask, "Cap_Status"] = np.nan

df.head()
```

```python
df["Cap_Status"] = df["Cap_Status"].str.strip().str.lower()

df["Cap_Status"] = df["Cap_Status"].replace({
    "capped": "Capped",
    "uncapped": "Uncapped"
})
```

```python
caps_columns = ["Test_Caps", "ODI_Caps", "T20_Caps", "IPL_Matches", "No_of_Matches"]

for col in caps_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
indian_mask = df["Country"] == "India"

df.loc[indian_mask, "State_Association"] = df.loc[indian_mask, "State_Association"].fillna(
    df.loc[indian_mask, "State_Association"].mode()[0]
)

df.loc[~indian_mask, "State_Association"] = df.loc[~indian_mask, "State_Association"].fillna("Not Applicable")
```

```python
num_cols = df.select_dtypes(include=np.number).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
def convert_price(price):
    if pd.isna(price):
        return np.nan

    price = str(price)
    number = re.findall(r"\d+\.?\d*", price)

    if len(number) == 0:
        return np.nan

    number = float(number[0])

    if "Cr" in price:
        return number * 100
    elif "Lakh" in price:
        return number
    else:
        return number
df["Reserve_Price_Numeric"] = df["Reserve_Price_Numeric"].fillna(
    df["Reserve_Price_Numeric"].median()
)
```

```python
print("Duplicate Rows Before:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicate Rows After:", df.duplicated().sum())
```

```python
Q1 = df["Reserve_Price_Numeric"].quantile(0.25)
Q3 = df["Reserve_Price_Numeric"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Reserve_Price_Numeric"] < lower) |
              (df["Reserve_Price_Numeric"] > upper)]

print("Number of Outliers Detected:", len(outliers))
```

```python
print("Final Shape:", df.shape)
print("\nRemaining Missing Values:")
print(df.isnull().sum())

df.to_csv("cricket_cleaned.csv", index=False)

print("Cleaned Dataset Saved Successfully")
```
