```python
df = pd.read_csv("cricket.csv")
```

```python
df = df.iloc[2:].reset_index(drop=True)
```

```python
df.columns = [
    "List_Sr_No",
    "Set_No",
    "Set_Code",
    "First_Name",
    "Surname",
    "Country",
    "State_Association",
    "Date_of_Birth",
    "Age",
    "Role",
    "Batting_Style",
    "Bowling_Style",
    "Test_Caps",
    "ODI_Caps",
    "T20_Caps",
    "IPL_Matches",
    "Previous_IPL_Teams",
    "Current_Team",
    "Jersey_No",
    "Cap_Status",
    "Reserve_Price"
]
```

```python
import numpy as np

df["Reserve_Price"] = pd.to_numeric(df["Reserve_Price"], errors="coerce")
```

```python
df["Reserve_Price"].isnull().sum()
```

```python
num_cols = [
    "Age",
    "Test_Caps",
    "ODI_Caps",
    "T20_Caps",
    "IPL_Matches",
    "Reserve_Price"
]

cat_cols = [
    "State_Association",
    "Bowling_Style",
    "Previous_IPL_Teams",
    "Current_Team",
    "Jersey_No",
    "Cap_Status"
]
```

```python
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(y=df["Reserve_Price"])
plt.title("Outlier Detection – Reserve Price")
plt.show()
```

```python
Q1 = df["Reserve_Price"].quantile(0.25)
Q3 = df["Reserve_Price"].quantile(0.75)
IQR = Q3 - Q1

outliers = df[
    (df["Reserve_Price"] < Q1 - 1.5 * IQR) |
    (df["Reserve_Price"] > Q3 + 1.5 * IQR)
]
```

```python
num_cols = ["Age", "Test_Caps", "ODI_Caps", "T20_Caps", "IPL_Matches"]

for col in num_cols:
    sns.boxplot(y=df[col])
    plt.title(f"Outlier Detection – {col}")
    plt.show()
```
